#  Download the House Pricing dataset from Kaggle and map the values to Aesthetics.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

# Load the California Housing dataset
california_housing = fetch_california_housing()
data = california_housing.data
feature_names = california_housing.feature_names
target = california_housing.target

# Create a DataFrame for easier manipulation
import pandas as pd
housing_df = pd.DataFrame(data, columns=feature_names)
housing_df['MedHouseVal'] = target  # Median House Value is the target variable

# Map values to aesthetics (Example: Scatter plot)
plt.figure(figsize=(12, 8))
sns.scatterplot(x='MedInc', y='MedHouseVal', hue='AveRooms', data=housing_df, alpha=1)
plt.title('Scatter Plot of Median Income vs Median House Value by Average Rooms')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

# Use different Color scales on the Rainfall Prediction dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'City': ['CityA', 'CityB', 'CityC', 'CityD'],
    'Jan': np.random.uniform(0, 50, 4),
    'Feb': np.random.uniform(0, 50, 4),
    'Mar': np.random.uniform(0, 50, 4),
    'April': np.random.uniform(0, 50, 4),
    'May': np.random.uniform(0, 50, 4),
    'June': np.random.uniform(0, 50, 4),
    'July': np.random.uniform(0, 50, 4),
    'Aug': np.random.uniform(0, 50, 4),
    'Sep': np.random.uniform(0, 50, 4),
    'Oct': np.random.uniform(0, 50, 4),
    'Nov': np.random.uniform(0, 50, 4),
    'Dec': np.random.uniform(0, 50, 4)
})


plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
sns.heatmap(df.set_index('City').iloc[:, 1:], annot = True, cmap = "cividis", cbar_kws = {'label': 'Rainfall'})

plt.subplot(2, 2, 2)
sns.heatmap(df.set_index('City').iloc[:, 1:], annot = True, cmap = "viridis", cbar_kws = {'label': 'Rainfall'})

plt.subplot(2, 2, 3)
sns.heatmap(df.set_index('City').iloc[:, 1:], annot = True, cmap = "BrBG", cbar_kws = {'label': 'Rainfall'})

plt.subplot(2, 2, 4)
sns.heatmap(df.set_index('City').iloc[:, 1:], annot = True, cmap = "coolwarm", cbar_kws = {'label': 'Rainfall'})

plt.tight_layout()
plt.show()
```

#  Create different Bar plots for variables in any dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset (built-in dataset in Seaborn)
iris = sns.load_dataset('iris')
print(iris)
# Bar plot for the average sepal length for each species
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='sepal_length', data=iris, palette = 'viridis',
                   capsize = 0.5, errwidth = 4, errcolor = 'c')
plt.title('Average Sepal Length for Each Species')
plt.show()

# Bar plot for the count of each species
plt.figure(figsize=(10, 6))
sns.countplot(x='species', data=iris)
plt.title('Count of Each Species')
plt.show()


# Stacked bar plot for the average sepal length and sepal width for each species
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='sepal_length', data=iris, ci=None, label='Sepal Length', color='blue')
sns.barplot(x='species', y='sepal_width', data=iris, ci=None, label='Sepal Width', color='orange')
plt.title('Average Sepal Length and Sepal Width for Each Species')
plt.legend()
plt.show()
```

# Show an example of Skewed data and removal of skewedness.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.special import boxcox1p

# Generate skewed data
np.random.seed(42)
skewed_data = np.random.exponential(scale=2, size=1000)

# Display the skewness of the original data
original_skewness = skew(skewed_data)
print(f"Original Skewness: {original_skewness}")

# Plot the original skewed data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(skewed_data, kde=True)
plt.title('Original Skewed Data')

# Apply Box-Cox transformation to reduce skewness
transformed_data = boxcox1p(skewed_data, 0.15)  # You can experiment with the parameter

# Display the skewness of the transformed data
transformed_skewness = skew(transformed_data)
print(f"Transformed Skewness: {transformed_skewness}")

# Plot the transformed data
plt.subplot(1, 2, 2)
sns.histplot(transformed_data, kde=True)
plt.title('Transformed Data (Reduced Skewness)')

plt.tight_layout()
plt.show()
```

# For a sales dataset do a Time Series visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a sales dataset with a 'date' column and a 'sales' column
# Replace the following lines with the actual loading of your sales dataset
# Example:
# sales_data = pd.read_csv('your_sales_data.csv')
# sales_data['date'] = pd.to_datetime(sales_data['date'])

# Creating a sample sales dataset for demonstration
date_rng = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')
sales_data = pd.DataFrame(date_rng, columns=['date'])
sales_data['sales'] = 100 + 5 * (date_rng.dayofweek) + 10 * np.random.randn(len(date_rng))

# Plotting the time series
plt.figure(figsize=(12, 6))
plt.plot(sales_data['date'], sales_data['sales'], label='Sales', color='blue')
plt.title('Sales Time Series')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.legend()
plt.grid(True)
plt.show()
```
