```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y)
plt.show()
```

```python
# linefmt

# "_"  -->   solid line
# "--" -->   dashed line
# "-." -->   dash-dot line
# ":"  -->   dotted line
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = "-.", markerfmt = "r", bottom = 0, basefmt = 'g', label = "python",
                use_line_collection = False, orientation = "vertical")

plt.legend()
plt.show()
```
