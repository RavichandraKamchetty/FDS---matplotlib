# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

plt.xlabel("Day", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("Scatter Plot", fontsize = 15)

plt.scatter(day, no, color = "r")
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Scatter Plot", fontsize=15)

colors = ["r", "y", "g", "b", "r", "g", "c"]
sizes = [400, 200, 400, 300, 200, 100, 600]

plt.scatter(day, no, c = colors, s = sizes, marker = "*", edgecolor = 'g', linewidth = 0.1)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no1 = [2, 3, 1, 4, 5, 3, 6]

plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Scatter Plot", fontsize=15)

colors = [10, 49, 30, 29, 56, 20, 30]
sizes = [400, 200, 400, 300, 200, 100, 600]


plt.scatter(day, no1, c = colors, s = sizes, cmap = "viridis", alpha = 0.5)
t = plt.colorbar()
t.set_label("Color Bar", fontsize = 15)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no1 = [2, 3, 1, 4, 5, 3, 6]
no2 = [3, 2, 4, 5, 1, 6, 2]

plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.title("Scatter Plot", fontsize=15)

colors = [10, 49, 30, 29, 56, 20, 30]
sizes = [400, 200, 400, 300, 200, 100, 600]


plt.scatter(day, no1, c = colors, s = sizes, cmap = "viridis")
plt.scatter(day, no2, c = 'r', s = sizes)

t = plt.colorbar()
t.set_label("Color Bar", fontsize = 15)
plt.show()
```
