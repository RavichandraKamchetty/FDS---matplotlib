# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["Python", "C", "C++", "Java"]
y = [75, 90, 80, 70]

plt.xlabel("Language", fontsize = 10)
plt.ylabel("No. ", fontsize = 10)
plt.title("RC", fontsize = 10)

c = ['y', 'b', 'm', 'g']
plt.bar(x, y, width = 0.5, color = c, align = "center", edgecolor = 'r', 
                            linewidth = 1, linestyle = ":", alpha = 0.7)
plt.show()
```

```python
x = ["Python", "C", "C++", "Java"]
y = [75, 90, 80, 70]

plt.xlabel("Language", fontsize = 10)
plt.ylabel("No. ", fontsize = 10)
plt.title("RC", fontsize = 10)

c = ['y', 'b', 'm', 'g']
plt.bar(x, y, width = 0.5, color = 'r', label = 'Popularity')
plt.legend()
plt.show()
```

```python
x = ["Python", "C", "C++", "Java"]
y = [75, 90, 80, 70]
z = [20, 30, 40, 50]

width = 0.3
p = np.arange(len(x))
p1 = [(j + width) for j in p]

plt.xlabel("Language", fontsize = 10)
plt.ylabel("No. ", fontsize = 10)
plt.title("RC", fontsize = 10)

plt.bar(p, y, width, color = 'r', label = 'Popularity1')
plt.bar(p1, z, width, color = 'y', label = 'Popularity2')

plt.xticks(p + (width / 2), x, rotation = 10)

plt.legend(loc = 1)
plt.show()
```

```python
x = ["Python", "C", "C++", "Java"]
y = [75, 90, 80, 70]
z = [20, 30, 40, 50]

width = 0.3
p = np.arange(len(x))
p1 = [(j + width) for j in p]

plt.xlabel("Language", fontsize = 10)
plt.ylabel("No. ", fontsize = 10)
plt.title("RC", fontsize = 10)

plt.barh(p, y, width, color = 'r', label = 'Popularity1')
plt.barh(p1, z, width, color = 'y', label = 'Popularity2')

plt.yticks(p + (width / 2), x, rotation = 0)

plt.legend(loc = 4)
plt.show()
```
