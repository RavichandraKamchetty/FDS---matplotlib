```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, notch = False, vert = False)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, labels = ["python"], patch_artist = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ["python"], showmeans = True, whis = 1.5, sym = "g+")
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ["python"], showmeans = True, boxprops = dict(color = "r"), 
                        capprops = dict(color = "m"), whiskerprops = dict(color = "g"),
                           flierprops = dict(markeredgecolor = "b"))
plt.show()
```

```python
x1 = [10, 20, 30, 40, 50, 60, 70]
x2 = [10, 20, 40, 50, 30, 60, 90]

y = [x1, x2]

plt.boxplot(y, labels = ["python", "c++"], showmeans = True, boxprops = dict(color = "r"), 
                        capprops = dict(color = "m"), whiskerprops = dict(color = "g"))
plt.show()
```
